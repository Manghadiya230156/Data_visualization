```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('penguins') 
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
df["species"].value_counts()
```

```python
df["island"].value_counts()
```

```python

sns.catplot(x="species", y="body_mass_g", hue="sex" , kind="bar", data=df)
plt.title("Average Body Mass by Species and Sex")
plt.show()
```

```python


c=sns.catplot(data=df, x="species", y="flipper_length_mm", kind="strip", col="sex",hue="sex")
c.fig.suptitle("Average Body Mass by Species and Sex", y=1.05)
```

```python
df1=df.loc[:, ['bill_length_mm','bill_depth_mm','flipper_length_mm','body_mass_g']]
cor_mat = df1.corr()
sns.heatmap(cor_mat, annot=True)
plt.title(" Correlation Heatmap of Penguin Features")
plt.show()
```

```python
sns.clustermap(cor_mat,linewidths=0.5,figsize=(7, 6),annot=True)
plt.suptitle("Clustered Correlation Map of Penguin Features", y=1.02)
plt.show()
```

```python
df2= pd.read_csv("tip.csv")
df2.head()
```

```python
df2.shape
```

```python
df2.info()
```

```python
df2.describe()
```

```python
sns.boxplot(x="day", y="total_bill", data=df2)
plt.title("Total Bill Distribution by Day")
plt.show()
```

```python
sns.violinplot(x="smoker", y="tip", data=df2)
plt.title("Tip Distribution by Smoker Status")
plt.show()
```

```python
sns.kdeplot(x='total_bill',hue='sex',data=df2,)
plt.title("Density of Total Bill by Gender")
```

```python

sns.jointplot(
data=df2,
x="total_bill",
y="tip",
hue="sex",
kind="scatter",
height=7
)
plt.suptitle("Total Bill vs Tip Relationship", y=1.02)
plt.show()
```

```python
sns.pairplot(df3,hue="species")
plt.suptitle("Pairwise Relationships in Iris Dataset", y=1.02)
plt.show()
```

```python
sns.swarmplot(x='species',y='petal_length',data=df3)
plt.title("Petal Length Distribution by Species")
```

```python
sns.scatterplot(data=df3,x="sepal_length", y="sepal_width", size="petal_length", hue="species",  alpha=0.7)
plt.title("Sepal vs Petal Dimensions in Iris")
plt.show()
```
