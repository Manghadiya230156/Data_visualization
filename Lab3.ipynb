```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,10,70)

m = 2
c = 0.3
y1 = 2 * x + 0.3
y2 = 2 * (x**2) + 0.3
y3 = 2 * (x**3) + 0.3
y4 = 2 * (x**0.5) + 0.3

plt.figure()
plt.plot(x, y1, label='y = 2x + 0.3')
plt.plot(x, y2, label='y = 2x² + 0.3')
plt.plot(x, y3, label='y = 2x³ + 0.3')
plt.plot(x, y4, label='y = 2√x + 0.3')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Line Chart: Linear, Quadratic, Square Root')
plt.legend()
plt.grid(True)
plt.show()
```

```python
x = np.linspace(-2,6,100)
y=(x*x) - (4*x) + 3 
plt.figure()
plt.plot(x, y, label='y=(x*x) - (4*x) + 3')

x_min = 2
y_min = (x_min)**2 - 4*x_min + 3
plt.scatter(x_min, y_min, color="red")
plt.annotate('min.point', xy=(x_min, y_min), xytext=(x_min+0.5, y_min+0.5),
             arrowprops=dict(facecolor='black',shrink=1))


plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2*np.pi, 100)

# Compute sin and cos
y_sin = np.sin(x)
y_cos = np.cos(x)

# Plot sin and cos
plt.figure()
plt.plot(x, y_sin, label='sin(x)')
plt.plot(x, y_cos, label='cos(x)')

plt.fill_between(x,y_sin,y_cos,where=(y_sin > y_cos),label='sin(x) > cos(x)')
plt.xlabel('x (radians)')
plt.ylabel('y')
plt.title('sin(x) and cos(x) from 0 to 2π')
plt.legend()
plt.show()
```

```python
x = np.linspace(0, 10, 200)

y1 = x**2
y2 = 2*x + 1
y3 = np.sqrt(x)

plt.figure()
plt.plot(x, y1, 'r--', label='y = x²')     
plt.plot(x, y2, 'b-', label='y = 2x + 1')    
plt.plot(x, y3, 'g:', label='y = √x') 

plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of Three Mathematical Functions')

plt.legend()
plt.grid(True)

plt.show()
```

```python
months = ["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"]
rainfall = [50, 40, 60, 80, 120, 200, 250, 220, 180, 100, 70, 55]
temperature = [15, 17, 20, 25, 30, 32, 31, 30, 28, 24, 20, 16]

fig, ax1 = plt.subplots(figsize=(10,6))

ax1.bar(months, rainfall, color='lightblue', label='Rainfall')
ax1.set_xlabel("Months")
ax1.set_ylabel("Rainfall", color='black')
ax1.tick_params(axis='y', labelcolor='black')

ax2 = ax1.twinx()
ax2.plot(months, temperature, color='red', marker='o', label='Temperature')
ax2.set_ylabel("Temperature", color='red')
ax2.tick_params(axis='y', labelcolor='red')

plt.title("Monthly Rainfall and Average Temperature")
plt.show()
```
